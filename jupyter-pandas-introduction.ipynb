The below line of code will import the `student-mat.csv` file as a pandas DataFrame object. Note that it specifies how the data is seperated (e.g., using a ';').

```python
dataframe = pd.read_csv('./student-data/student-mat.csv', sep=';')
```

The `.head()` command will show the first 5 lines of the DataFrame.

```python
dataframe.head()
```

The `.columns` attribute returns a list of column names.

```python
dataframe.columns
```

The `.describe()` method returns another DataFrame with descriptive statistics on all of the numerical columns, including the average, quartiles and min/max.

```python
dataframe.describe()
```

Here, I'm indexing into just the 'age' column of my DataFrame and calling the `.hist()` method to plot a histogram.

```python
dataframe['age'].hist()
```

Below I'm indexing into my DataFrame to only get the rows where the 'sex' of the student is 'F'.

```python
dataframe[dataframe['sex'] == 'F']
```

Here's how I get the average age of students who were in a relationship.

```python
dataframe[dataframe['romantic'] == 'yes']['age'].mean()
```
